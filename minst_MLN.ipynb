```python
# date: 2018/07/03
# Construct a MLN network to solve the mnist handwriting problem
# Main layers included in the MLN network: 
# Hidden layer 
# Output layer
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
np.random.seed(10)
```

```python
from keras.datasets import mnist
```

```python
(x_train_image, y_train_label), \
(x_test_image, y_test_label) = mnist.load_data()
```

```python
print('train data=', x_train_image.shape)
print('test data=', y_train_label.shape)
```

```python
import matplotlib.pyplot as plt
def plot_image(image):
    fig = plt.gcf()
    fig.set_size_inches(2,2)
    plt.imshow(image, cmap = 'binary')
    plt.show()
```

```python
x_train = x_train_image.reshape(60000, 784).astype('float32')
x_test = x_test_image.reshape(10000, 784).astype('float32')
```

```python
print('x_train_image', x_train.shape)
print('x_test_image', x_train.shape)
```

```python
x_train[0]
```

```python
x_train_normalize = x_train/255
x_test_normalize = x_test/255
```

```python
x_train_normalize[0]
```

```python
y_train_label[:5]
```

```python
y_trainonehot = np_utils.to_categorical(y_train_label)
y_testonehot = np_utils.to_categorical(y_test_label)
```

```python
y_trainonehot[:5]
```

```python
model = Sequential()
```

```python
#Construct hidden layer
model.add(Dense(units = 256,
                input_dim = 784,
                kernel_initializer = 'normal',
                activation = 'relu'
))
```

```python
#Construct output layer
model.add(Dense(units = 10,
                kernel_initializer = 'normal',
                activation = 'softmax'
))
```

```python
print(model.summary())
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam', metrics = ['accuracy']
)
```

```python
train_history = model.fit(x = x_train_normalize,
                          y = y_trainonehot, validation_split = 0.2,
                          epochs = 10, batch_size = 200, verbose = 2
                         )
```
